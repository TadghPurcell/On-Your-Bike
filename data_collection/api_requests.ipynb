```python
import requests
import json
from pprint import pprint

# Get the API key for Dublin bikes
f = open("./dbinfo.json")
BIKE_API_KEY = json.load(f)['JCKey']
f.close()

#Get the API key for Open weather
f = open("./dbinfo.json")
WEATHER_API_KEY = json.load(f)['weatherKey']
f.close()

# URI and name
STATIONS_URI = 'https://api.jcdecaux.com/vls/v1/stations'
NAME = 'dublin'
#Weather URI 
WEATHER_URI = 'https://api.openweathermap.org/data/2.5/weather'
```

```python
# Make an API request
r = requests.get(STATIONS_URI, params={'apiKey':BIKE_API_KEY, 'contract':NAME})
json.loads(r.text)
```

```python
# Moby API
r = requests.get("https://data.smartdublin.ie/mobybikes-api/last_reading/")
json.loads(r.text)
```

```python
# Bleeper
r = requests.get("https://data.smartdublin.ie/bleeperbike-api/last_snapshot/")
json.loads(r.text)
```

```python
#Weather 
w = requests.get(WEATHER_URI, params={"units": "metric", "lat": 53.344, "lon": -6.2672, "appid": WEATHER_API_KEY})
json.loads(w.text)
```
