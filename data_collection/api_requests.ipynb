```python
import requests
import json
from pprint import pprint

# Get the API key for Dublin bikes
f = open("dbinfo.json")
API_KEY = json.load(f)['JCKey']

# URI and name
STATIONS_URI = 'https://api.jcdecaux.com/vls/v1/stations'
NAME = 'dublin'
```

```python
# Make an API request
r = requests.get(STATIONS_URI, params={'apiKey':API_KEY, 'contract':NAME})
json.loads(r.text)
```

```python
# Moby API
r = requests.get("https://data.smartdublin.ie/mobybikes-api/last_reading/")
json.loads(r.text)
```

```python
# Bleeper
r = requests.get("https://data.smartdublin.ie/bleeperbike-api/last_snapshot/")
json.loads(r.text)
```
